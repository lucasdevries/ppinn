```python
import os, glob
import pydicom
from datetime import datetime
```

```python
ctp_dir = r'L:\basic\divi\CMAJOIE\CLEOPATRA\CTP_MRI_data\Data\CTP_DICOM'
mri_dir = r'L:\basic\divi\CMAJOIE\CLEOPATRA\CTP_MRI_data\Data\MRI_DICOM'
PIDs = ['C'+str(x).zfill(3) for x in range(102,117)]
```

```python
def days_hours_minutes(td):
    return td.days, td.seconds//3600, (td.seconds//60)%60, 60*(td.seconds//3600) + (td.seconds//60)%60

for ID in PIDs:
    ctp_folder = os.path.join(ctp_dir, ID)
    mri_folder = os.path.join(mri_dir, ID+'*')
    ctp_file = glob.glob(os.path.join(ctp_folder, '**/*.dcm'), recursive=True)[0]
    mri_file = glob.glob(os.path.join(mri_folder, '**/*.dcm'), recursive=True)[0]

    ctp_ds = pydicom.dcmread(ctp_file)
    mri_ds = pydicom.dcmread(mri_file)
    # print(ID, ctp_ds[0x0008, 0x1010].value, mri_ds[0x0008, 0x1010].value)
    if ID == 'C109':
        print(ID, ctp_ds, mri_ds)


    try:
        time_ctp = datetime.strptime(ctp_ds[0x0008, 0x0031].value, "%H%M%S.%f")
    except:
        time_ctp = datetime.strptime(ctp_ds[0x0008, 0x0031].value, "%H%M%S")
    try:
        time_mri = datetime.strptime(mri_ds[0x0008, 0x0032].value, "%H%M%S.%f")
    except:
        time_mri = datetime.strptime(mri_ds[0x0008, 0x0032].value, "%H%M%S")
    #
    print(ID, days_hours_minutes(time_mri-time_ctp))
    # print(ID, time_mri.minute -time_ctp)
```
